In this notebook we briefly present the models used in this study.

They are custom written ResNet34, ResNet50 and VGG19 architectures such that:

- The convolutional part ends in a Conv2d layer

- The GAP layer is replaced by a Flatten layer

```python
from resnet import ResNet
from vgg19 import VGG19
```

##### Without loss of generality we consider 10 classes. This can be adjusted to the dataset of interest.

## ResNet34

```python
resnet34 = ResNet(type_='34',
                 no_of_classes=10,
                 trainable_layers=None)
resnet34.model
```

## ResNet50

```python
resnet50 = ResNet(type_='50',
                 no_of_classes=10,
                 trainable_layers=None)
resnet50.model
```

## VGG19

```python
vgg19 = VGG19(num_classes=10,
              trainable_feature_layers=None)
vgg19.model
```
